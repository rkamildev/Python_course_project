```python
import pandas as pd
import numpy as np
import datetime
from matplotlib import pyplot as plt
from matplotlib.pyplot import plot

df = pd.read_csv('e:\Course_Project/big_five_data/big_five_stocks.csv') #path can differ
print(df.head())
```

```python
df.name.unique()
```

```python
#Adjusting date time
df["Unnamed: 0"] = pd.to_datetime(df["Unnamed: 0"])
df["year"] = df["Unnamed: 0"].dt.year
df["month"] = df["Unnamed: 0"].dt.month
df = df.rename({"Unnamed: 0" : "date"}, axis = 1, inplace = False)
```

```python
#getting stocks by the name
aapl = df[df.name == "AAPL"].drop("name", axis = 1)
fb = df[df.name == "FB"].drop("name", axis = 1)
msft = df[df.name == "MSFT"].drop("name", axis = 1)
amzn = df[df.name == "AMZN"].drop("name", axis = 1)
googl = df[df.name == "GOOGL"].drop("name", axis = 1)


aapl.set_index("date", drop = True, inplace = True)
fb.set_index("date", drop = True, inplace = True)
msft.set_index("date", drop = True, inplace = True)
amzn.set_index("date", drop = True, inplace = True)
googl.set_index("date", drop = True, inplace = True)
```

```python
#plotting opening and closing stocks values 

#Apple
plt.figure(figsize = (20,6))

aapl["close"].plot(color = "red", alpha = 2, label = "close")
aapl["open"].plot(color = "blue", label = "open")


plt.title("Apple", fontsize = 16)
plt.legend()
plt.show()

#Facebook
plt.figure(figsize = (20,6))

fb["close"].plot(color = "red", label = "close")
fb["open"].plot(color = "blue", alpha = 1, label = "open")

plt.title("Facebook", fontsize = 16)
plt.legend()
plt.show()

#Microsoft
plt.figure(figsize = (20,6))

msft["close"].plot(color = "red", label = "close")
msft["open"].plot(color = "blue", alpha = 1, label = "open")

plt.title("Microsoft", fontsize = 16)
plt.legend()
plt.show()

#Amazon
plt.figure(figsize = (20,6))

amzn["close"].plot(color = "red", label = "close")
amzn["open"].plot(color = "blue", alpha = 1, label = "open")

plt.title("Amazon", fontsize = 16)
plt.legend()
plt.show()


#Google
plt.figure(figsize = (20,6))

googl["close"].plot(color = "red", label = "close")
googl["open"].plot(color = "blue", alpha = 1, label = "open")

plt.title("Google", fontsize = 16)
plt.legend()
plt.show()
```

```python
#bar chart with last open and close values comparison
w = 0.4
plt.figure(figsize = (15, 10))
x = ["Apple", "Facebook", "Microsoft", "Amazon", "Google"]

openl_aapl = aapl["open"].values[-1]
closel_aapl = aapl["close"].values[-1]

openl_fb = fb["open"].values[-1]
closel_fb = fb["close"].values[-1]

openl_msft = msft["open"].values[-1]
closel_msft = msft["close"].values[-1]

openl_amzn = amzn["open"].values[-1]
closel_amzn = amzn["close"].values[-1]


openl_googl = googl["open"].values[-1]
closel_googl = googl["close"].values[-1]

open_v = [openl_aapl, openl_fb, openl_msft, openl_amzn, openl_googl]
close_v = [closel_aapl, closel_fb, closel_msft, closel_amzn, closel_googl]

bar1 = np.arange(len(x))
bar2 = [i+w for i in bar1]

plt.bar(bar1, open_v, w, label = "Open Value")
plt.bar(bar2, close_v, w, label = "Close Value")

plt.xlabel("Companies", fontsize = 16)
plt.ylabel("Values")
plt.title("Last Open&Close values", fontsize = 16)
plt.xticks(bar1+w/2, x)
plt.legend()
plt.show()
```

```python
#The volume of bought stocks in millions

#Apple
plt.figure(figsize = (20,6))

aapl["volume"].plot(color = "green", label = "close")

plt.title("Apple", fontsize = 16)
plt.legend()
plt.show()

#Facebook
plt.figure(figsize = (20,6))

fb["volume"].plot(color = "green", label = "close")

plt.title("Facebook", fontsize = 16)
plt.legend()
plt.show()

#Microsoft
plt.figure(figsize = (20,6))

msft["volume"].plot(color = "green", label = "close")

plt.title("Microsoft", fontsize = 16)
plt.legend()
plt.show()

#Amazon
plt.figure(figsize = (20,6))

amzn["volume"].plot(color = "green", label = "close")

plt.title("Amazon", fontsize = 16)
plt.legend()
plt.show()

#Google
plt.figure(figsize = (20,6))

googl["volume"].plot(color = "green", label = "close")

plt.title("Google", fontsize = 16)
plt.legend()
plt.show()

```

```python
#Percentage of the all stocks bought

max_aapl = aapl["volume"].max()    
max_fb = fb["volume"].max()
max_msft = msft["volume"].max()
max_amzn = amzn["volume"].max()
max_googl = googl["volume"].max()

min_aapl = aapl["volume"].min()    
min_fb = fb["volume"].min()
min_msft = msft["volume"].min()
min_amzn = amzn["volume"].min()
min_googl = googl["volume"].min()

avg_aapl = (max_aapl + min_aapl) / 2
avg_fb = (max_fb + min_fb) / 2
avg_msft = (max_msft + min_msft) / 2
avg_amzn = (max_amzn + min_amzn) / 2
avg_googl = (max_googl + min_googl) / 2


sum = avg_aapl + avg_fb + avg_msft + avg_amzn + avg_googl

aapl_r = sum / avg_aapl
fb_r = sum / avg_fb
msft_r = sum / avg_msft
amzn_r = sum / avg_amzn
googl_r = sum / avg_googl


labels = 'Apple', 'Facebook', 'Microsoft', 'Amazon', 'Google'
sizes = [aapl_r, fb_r, msft_r, amzn_r, googl_r]

plt.figure(figsize = (10, 10))
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
#Growth percentage

lowest_aapl = aapl["low"].values[0]
lowest_fb = fb["low"].values[0]
lowest_msft = msft["low"].values[0]
lowest_amzn = amzn["low"].values[0]
lowest_googl = googl["low"].values[0]

highest_aapl = aapl["high"].values[-1]
highest_fb = fb["high"].values[-1]
highest_msft = msft["high"].values[-1]
highest_amzn = amzn["high"].values[-1]
highest_googl = googl["high"].values[-1]

#Growth precentage calculation

grp_aapl = highest_aapl / lowest_aapl
grp_fb = highest_fb  / lowest_fb
grp_msft = highest_msft / lowest_msft
grp_amzn = highest_amzn / lowest_amzn
grp_googl = highest_googl / lowest_googl

#Growth display
plt.rcdefaults()
fig, ax = plt.subplots()
# Example data
companies = ('Apple', 'Facebook', 'Microsoft', 'Amazon', 'Google')
y_pos = np.arange(len(companies))
percentile = [grp_aapl, grp_fb, grp_msft, grp_amzn, grp_googl]

ax.barh(y_pos, percentile)
ax.set_yticks(y_pos)
ax.set_yticklabels(companies)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Percentile')
ax.set_title('Growth percentage')

plt.show()
```
